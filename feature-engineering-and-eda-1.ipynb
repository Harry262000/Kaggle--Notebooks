```python
train_df = pd.read_excel("/kaggle/input/flight-price/Data_Train.xlsx")
train_df.head()

```

```python
test_df = pd.read_excel("/kaggle/input/flight-price/Test_set.xlsx")
test_df.head()
```

```python
df = train_df.append(test_df)
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
# Method - 1
df['Date_of_Journey'].str.split("/").str[0]
```

```python
#  Date of journey column was in object and make split (DAY, MONTH, YEAR)
# Feature Engineering steps
df['Date'] = df['Date_of_Journey'].str.split("/").str[0]
df['Month'] = df['Date_of_Journey'].str.split("/").str[1]
df['Year'] = df['Date_of_Journey'].str.split("/").str[2]
```

```python
df.head(20)
```

```python
# Method - 2 , TO deal with features using lambda function

df['Date'] = df['Date_of_Journey'].apply(lambda x:x.split("/")[0])
df['Month'] = df['Date_of_Journey'].apply(lambda x:x.split("/")[1])     
df['Year'] = df['Date_of_Journey'].apply(lambda x:x.split("/")[2])
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.drop(['Date_of_Journey'], axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
df['Arrival_Time'].str.split(' ').str[0]
```

```python
# for split(' '), having empty space Method one doesn't work all time, but lambda works
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(" ")[0])
```

```python
df.head(2)
```

```python
df['Arrival_hour'] = df['Arrival_Time'].apply(lambda x:x.split(":")[0])
df['Arrival_min'] = df['Arrival_Time'].apply(lambda x:x.split(":")[1])
```

```python
df.head(2)
```

```python
df.drop(['Arrival_Time'], axis=1, inplace=True)
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
```

```python
# split dept_time to hour , min

df['Dep_hour'] = df['Dep_Time'].apply(lambda x:x.split(":")[0])
df['Dep_min'] = df['Dep_Time'].apply(lambda x:x.split(":")[1])

#dept_time object to int , so we could use it in our visualization

```

```python
df['Dep_hour'] = df['Dep_hour'].astype(int)
df['Dep_min'] = df['Dep_min'].astype(int)

# dept_hour , drop

df.drop("Dep_Time", axis=1, inplace=True)
```

```python
df.info()
```

```python
df['Total_Stops'].fillna(df['Total_Stops'].mode()[0], inplace=True)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop': 0 , '2 stops': 2, '1 stop': 1, '3 stops' : 3, '4 stops': 4 })
```

```python
df.drop('Route', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df['Additional_Info'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
df['Airlines'] = labelencoder.fit_transform(df['Airline'])
df['Source'] = labelencoder.fit_transform(df['Source'])
df['Destination'] = labelencoder.fit_transform(df['Destination'])
df['Additional_info'] = labelencoder.fit_transform(df['Additional_Info'])
```
